```python
import os
import numpy as np

DATA_PATH = '/home/yadanar/Desktop/School Work/Advance Topic in Global Software Engineering/Vietnamese-Sign-Language-Translation/Sign Language Translator/Data'

print("üìä COMPLETE DATASET SIZE CHECK")
print("=" * 60)

# Count everything
action_folders = os.listdir(DATA_PATH)
print(f"1. Actions (different signs): {len(action_folders)}")

# Check first action
first_action = os.path.join(DATA_PATH, action_folders[0])
sequences = os.listdir(first_action)
print(f"2. Sequences in first action: {len(sequences)}")

# Check first sequence
first_sequence = os.path.join(first_action, sequences[0])
frames = [f for f in os.listdir(first_sequence) if f.endswith('.npy')]
print(f"3. Frames in first sequence: {len(frames)}")

# Load multiple frames to see REAL ranges
print(f"\nüîç ANALYZING MULTIPLE FRAMES FOR TRUE RANGES:")

x_values, y_values, z_values = [], [], []
for frame_file in frames[:10]:  # Check first 10 frames
    data = np.load(os.path.join(first_sequence, frame_file))
    landmarks = data.reshape(21, 6)
    x_values.extend(landmarks[:, 0])
    y_values.extend(landmarks[:, 1])
    z_values.extend(landmarks[:, 2])

print(f"X range across 10 frames: [{min(x_values):.3f}, {max(x_values):.3f}]")
print(f"Y range across 10 frames: [{min(y_values):.3f}, {max(y_values):.3f}]")
print(f"Z range across 10 frames: [{min(z_values):.3f}, {max(z_values):.3f}]")

# Calculate totals
total_frames = 0
for action in action_folders[:3]:  # Check first 3 actions only
    action_path = os.path.join(DATA_PATH, action)
    sequences = os.listdir(action_path)
    for seq in sequences[:3]:  # First 3 sequences only
        seq_path = os.path.join(action_path, seq)
        frames = [f for f in os.listdir(seq_path) if f.endswith('.npy')]
        total_frames += len(frames)

print(f"\nüìà PROJECTED TOTALS (if all actions have same):")
print(f"Actions: {len(action_folders)}")
print(f"Sequences per action: ~{len(os.listdir(first_action))}")
print(f"Frames per sequence: ~{len([f for f in os.listdir(first_sequence) if f.endswith('.npy')])}")
print(f"Total frames: ~{len(action_folders) * len(sequences) * len(frames):,}")
print(f"Total data points: ~{len(action_folders) * len(sequences) * len(frames) * 126:,}")
```
